```python
import matplotlib.pyplot as plt
import numpy as np

dados = {
    "ano": [2018, 2019, 2020, 2021, 2022, 2023, 2024],
    "brasil": [843, 942, 1070, 1230, 1420, 1640, 1900],
    "eua": [139000, 152000, 177000, 212000, 258000, 318000, 395000],
    "portugal": [2500, 2700, 3100, 3900, 4900, 6200, 7800],
    "finlandia": [1100, 1200, 1300, 1500, 1800, 2100, 2600],
    "suecia": [1800, 2200, 2700, 3500, 4500, 5700, 7200],
    "alemanha": [27700, 28900, 32300, 38100, 46700, 58100, 72800],
}

# Criar um novo gráfico
plt.figure()

# Normalizar os valores dos países (exceto 'ano')
for key in dados:
    if key != 'ano':
        dados[key] = [(x - min(dados[key])) /
                      (max(dados[key]) - min(dados[key])) for x in dados[key]]

# Plotar as séries normalizadas
for key in dados:
    if key != 'ano':
        plt.plot(dados['ano'], dados[key], label=key)

# Adicionar legendas, rótulos e título
plt.legend()
plt.xlabel('Ano')
plt.ylabel('Valores Normalizados')
plt.title('Gráfico de Linhas com Valores Normalizados')

plt.savefig('plot_crescimento.png')

# Mostrar o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

dados = {
    "ano": [2018, 2019, 2020, 2021, 2022, 2023, 2024],
    "brasil": [843, 942, 1070, 1230, 1420, 1640, 1900],
    "eua": [139000, 152000, 177000, 212000, 258000, 318000, 395000],
    "portugal": [2500000, 2700000, 3100000, 3900000, 4900000, 6200000, 7800000],
    "finlandia": [1100000, 1200000, 1300000, 1500000, 1800000, 2100000, 2600000],
    "suecia": [1800000, 2200000, 2700000, 3500000, 4500000, 5700000, 7200000],
    "alemanha": [27700000, 28900000, 32300000, 38100000, 46700000, 58100000, 72800000],
}

# Criar um novo gráfico
plt.figure()

# Normalizar os valores dos países (exceto 'ano') usando Z-score
for key in dados:
    if key != 'ano':
        dados[key] = [(x - np.mean(dados[key])) / np.std(dados[key])
                      for x in dados[key]]

# Plotar as séries normalizadas
for key in dados:
    if key != 'ano':
        plt.plot(dados['ano'], dados[key], label=key)

# Adicionar legendas, rótulos e título
plt.legend()
plt.xlabel('Ano')
plt.ylabel('Valores Normalizados (Z-score)')
plt.title('Gráfico de Linhas com Valores Normalizados')

plt.savefig('plot_crescimento.png')
# Mostrar o gráfico
plt.show()
```
