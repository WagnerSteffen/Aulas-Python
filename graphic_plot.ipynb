```python
import matplotlib.pyplot as plt

# Dados do IDH do Brasil e de Cuba
anos = [1960, 1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020]
idh_brasil = [0.523, 0.545, 0.578, 0.611, 0.628, 0.649, 0.665, 0.688, 0.705, 0.725, 0.730, 0.754, 0.765]
idh_cuba = [0.651, 0.669, 0.718, 0.729, 0.733, 0.732, 0.726, 0.733, 0.738, 0.776, 0.780, 0.777, 0.778]

# Cria o gráfico em linha
plt.plot(anos, idh_brasil, label='Brasil')
plt.plot(anos, idh_cuba, label='Cuba')

# Configurações do gráfico
plt.title('IDH do Brasil e de Cuba ao longo do tempo')
plt.xlabel('Ano')
plt.ylabel('IDH')
plt.legend()

# Mostra o gráfico
plt.show()
```
